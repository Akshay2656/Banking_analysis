### Descriptive analysis

```python
df.describe()
```

```python
range = [0, 100000, 300000 , float('inf')]
labels = ["low", "Mediam", "High"]
df["Income Band"] = pd.cut(
    df['Estimated Income'],
    bins = range,
    labels = labels,
    right = False
)
```

```python
df['Income Band'].value_counts()
```

```python
df['Income Band'].value_counts().plot(kind = 'bar')
```

```python
# examine the distribution of unique chategories in categorical columns 
categorical_columns = df[["BRId",
                          "GenderId",
                          "IAId",
                          "Amount of Credit Cards",
                          "Nationality",
                          "Occupation",
                          "Fee Structure",
                          "Loyalty Classification",
                          "Properties Owned",
                          "Risk Weighting",
                          "Income Band"
                         ]].columns
for column in categorical_columns:
    print(f"value counts for \"{column}\":")
    display(df[column].value_counts())
    
```

```python
for column in categorical_columns:
    plt.figure(figsize=(10,6))
    sns.countplot(data=df, x=column, hue=column, palette='pastel', legend=False)
    plt.title(f'Value counts for "{column}"')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Histplot of value count for different occupation

for column in categorical_columns:
    if column == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[column])
    plt.title("Histogram of Occupation count")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.show()
```

### Numerical Analysis

```python
numerical_columns = ['Estimated Income',
                  'Superannuation Savings',
                  'Credit Card Balance',
                  'Bank Loans',
                  'Bank Deposits',
                  'Checking Accounts',
                  'Saving Accounts',
                  'Foreign Currency Account',
                  'Business Lending'
                 ]

# Univariant analysis and visualization
plt.figure(figsize=(16,15))
for i, column in enumerate(numerical_columns):
    plt.subplot(4,3,i+1)
    plt.subplots_adjust(hspace=0.4, wspace=0.3)
    sns.histplot(df[column], kde=True)
    plt.title(column)
plt.show()
```

```python
numerical_columns = ['Estimated Income',
                  'Superannuation Savings',
                  'Credit Card Balance',
                  'Bank Loans',
                  'Bank Deposits',
                  'Checking Accounts',
                  'Saving Accounts',
                  'Foreign Currency Account',
                  'Business Lending'
                 ]
correlation_matrix = df[numerical_columns].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot= True, cmap= "crest", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## Insight Of EDA 

#### The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.
